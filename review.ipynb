```python
s="hello"
def reversestring(s):
    if len(s)==0:
        return s
    else:
        return s[-1]+reversestring(s[:-1])

print(reversestring(s))
```

```python
arr=[1,2,3,4,5]
l=len(arr)
def sum(arr,l):
    if l<=0:
        return 0
    else:
        return arr[l-1]+sum(arr,l-1)
    
print(sum(arr,l))
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.nref=None
        self.pref=None

class Linkedlist:
    def __init__(self):
        self.head=None

    def insert(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
            print(self.head.data)
        else:
            n=self.head
            while n.nref is not None:
                n=n.nref
            new_node.pref=n
            n.nref=new_node

    def printll(self):
        n=self.head
        while n  is not None:
            print(n.data,end="->")
            n=n.nref
        print("none")

l=Linkedlist()
l.insert(10)
l.insert(20)
l.printll()
```
