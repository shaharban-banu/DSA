```python
arr=[1,3,5,7,9]
tar=15
# create a function to check tar exist in arr
# input : arr, tar 
# ouput :true or False
# use less complex search operation that can be applicable on array

def search(arr,tar):
    low=0
    high=len(arr)-1
    while low<=high:
        mid=(low+high)//2
        if arr[mid]==tar:
            return True
        if arr[mid]<tar:
            low=mid+1
        else:
            high=mid-1
    return False
print(search(arr,tar))
```

```python
arr=[1,2,6,3,4,5]

def largest(arr):
    if len(arr)==0:
        return 0
    return max(arr[-1],largest(arr[:-1]))
print(largest(arr))
```
