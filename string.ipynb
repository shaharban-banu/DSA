```python
#Check if Two Strings are Anagrams.
#Two strings are anagrams if they contain the same characters in the same count, just possibly in a different order.
s="listen"
t="nestil"
if sorted(s)==sorted(t):
    print("anagram")
else:
    print("not anagram")

```

```python
from collections import Counter
if Counter(s)==Counter(t):
    print("anagram")
else: print("not")
```

```python
if len(s)!=len(t):
    print("not anagram")
count1={}
count2={}
for c in s:
    if c  in count1:
        count1[c]+=1
    else:
        count1[c]=1

for c in t:
    if c in count2:
        count2[c]+=1
    else:
        count2[c]=1
if count1==count2:
    print("anagram")
else:
    print("not anagram")
```

```python
#Replace each alphabet in the string with the n-th character after it in the alphabet.
s="h2ello3BYE"
n=3
result=""
for char in s:
    if char.isalpha():
        base=ord('a') if char.islower() else ord('A')
        shift=chr((ord(char)-base+n) % 26 +base)
        result+=shift
    else:
        result+=char
print(result)
```

```python
#palindrome
s="A man, a plan, a canal: Panama"
s=" ".join(c.lower() for c in s if c.isalnum())
print(s)
if s==s[::-1]:
    print("palindrome")
```

```python
#palindrome
s="A man, a plan, a canal: Panama"
left, right = 0, len(s) - 1

while left < right:
    # Move left to the next alphanumeric character
    while left < right and not s[left].isalnum():
        left += 1
    # Move right to the previous alphanumeric character
    while left < right and not s[right].isalnum():
        right -= 1
    # Compare lowercase versions
    if s[left].lower() != s[right].lower():
        print("False")
    left += 1
    right -= 1
print("True")
   
```

```python
#Compress a String (Basic Run-Length Encoding)
#E.g. "aaabbc" → "a3b2c1"
s="aaabbcc"
r=""
count=1
for i in range(1,len(s)):
    if s[i]==s[i-1]:
        count+=1
    else:
        r+=s[i-1]+str(count)
        count=1
r+=s[i-1]+str(count)
print(r)
```

```python
def permute(s, answer=""):
    if len(s) == 0:
        print(answer)
        return
    for i in range(len(s)):
        ch = s[i]
        left = s[:i]
        right = s[i+1:]
        permute(left + right, answer + ch)

permute("abc")
```
