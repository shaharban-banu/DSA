```python
#movezeroes
nums = [0,1,0,3,12]
j=0
for i in range (len(nums)):
    if nums[i]!=0:
        nums[j]=nums[i]
        j+=1
while j<len(nums):
    nums[j]=0
    j+=1
print(nums)
```

```python
#278. First Bad Version
n = 10
bad = 4

def isBadVersion(x):
    if x>=bad:
        return True
    else :
        return False
left=0
right=n
while left<right:
    mid=(left+right)//2
    if isBadVersion(mid):
        right=mid
    else:
        left=mid+1

print(left)
```
