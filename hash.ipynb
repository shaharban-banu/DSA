```python
#list implementantion and chaining to handle collision
table_size=10
hash_table=[[]for _ in range(table_size)]

def hashing(key):
    return hash(key)%table_size

def insert(key,value):
    index=hashing(key)
    for item in hash_table[index]:
        if item[0]==key:#if key alredy present update it
            item[1]=value
            return
    hash_table[index].append([key,value])#otherwise add it to hash table

def display():
    for i,item in enumerate( hash_table):
        print("bucket",i,":",item)

def search(key):
    index=hashing(key)
    for item in hash_table[index]:
        if item[0]==key:
            return item[1]
    return None

def delete(key):
    index=hashing(key)
    for i,item in enumerate(hash_table[index]):
        if item[0]==key:
            del hash_table[index][i]
            return True
    return False

display()
insert('a','apple')
insert('b','bat')
insert('c','cat')
print("after insertion")
display()

print("value of key :",search('b'))
delete('b')
```

#open addressing

```python
table_size=10
hash_table=[None]*table_size
DELETED=object()#marker used to represent deleted slot

def hashing(key):
    return hash(key)%table_size

def insert(key,value):
    index=hashing(key)
    original_index=index
    while hash_table[index] is not None and hash_table[index] is not DELETED:
        if hash_table[index][0]==key:
            hash_table[index][1]=value
            return
        index=(index+1)%table_size#linear probing
        if index==original_index:#no empty slot
            raise Exception("hash table is full")#u can just print it and return
    hash_table[index]=[key,value]

def display():
    for i ,item in enumerate (hash_table):
        if item is DELETED:
            print("index",i,": DELETED")
        else:
            print("index",i,':',item)

def search(key):
    index=hashing(key)
    original_index=index
    while hash_table[index] is not None:
        if hash_table[index] is not DELETED and hash_table[index][0]==key:
            return hash_table[index][1]
        index=(index+1)%table_size
        if index==original_index:#no empty slot
            break
    return False

def delete(key):
    index=hashing(key)
    original_index=index
    while hash_table[index] is not None:
        if hash_table[index] is not DELETED and hash_table[index][0]==key:
            hash_table[index]=DELETED
            return True
        index=(index+1)%table_size
        if index==original_index:
            break
    return None

display()
insert('a','apple')
insert('b','bat')
insert('c','cat')
print("------------------")
print("after insertion")
display()
print("------------------")
print("value of key a:",search('a'))
print("------------------")
delete('a')
display()
print("------------------")
```
