```python
#bubble sort
# arr=[1,2,3,2,4]
arr=[]
num=int(input("how many number you want to enter :"))
print("enter values :")
for _ in range(num):
    arr.append(int(input()))

for i in range(len(arr)-1):
    for j in range(len(arr)-1-i):#-i -->to avoid compairing elementwith the already sorted element
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
print("sorted array ;",arr)
```

selection sort using min/max method

```python
#selection sort for non duplicate numbers
arr=[56,3,2,78,6,0]
print(arr)

for i in range(len(arr)-1):
        min_value=min(arr[i:])#i want ith smallest element in ith  iteration
        min_indx=arr.index(min_value)
        arr[min_indx],arr[i]=arr[i],arr[min_indx]
           
print(arr)
```

```python
#selection sort for duplicate numbers
arr=[56,3,2,78,2,0,6,0]
print(arr)

for i in range(len(arr)-1):
        min_value=min(arr[i:])
        min_indx=arr.index(min_value,i)
        if arr[i]!=arr[min_indx]:
            arr[min_indx],arr[i]=arr[i],arr[min_indx]
           
print(arr)
```

selection sort without min/max method-
take first element as min or max

```python
#take 1st element as min value
for i in range(len(arr)-1):
    m_val=arr[i]
    for j in range(i+1,len(arr)):
        if arr[j]<m_val:
            m_val=arr[j]
    m_indx=arr.index(m_val,i)
    if arr[i]!=arr[m_indx]:
        arr[i],arr[m_indx]=arr[m_indx],arr[i]
print(arr)
```

```python
#without using index method
# take 1st index  as min index
num=int(input("how many numbers :"))
arr=[int(input("enter number :")) for x in range(num)]
print("unsorted array :",arr)
for i in range(len(arr)-1):
    m_indx=i
    for j in range(i+1,len(arr)):
        if arr[j]<arr[m_indx]:
            m_indx=j
    
    if arr[i]!=arr[m_indx]:
        arr[i],arr[m_indx]=arr[m_indx],arr[i]
print(arr)
```

```python
#insertion sort
arr=[2,4,3,5,1]
for i in range(1,len(arr)):
    pos=i
    c_val=arr[i]
    while c_val<arr[pos-1] and pos>0:
         arr[pos]=arr[pos-1]
         pos-=1
    arr[pos]=c_val
print(arr)
```

```python
def merge_sort(arr):

    if len(arr)>1:
        mid=len(arr)//2
        left=arr[:mid]
        right=arr[mid:]
        merge_sort(left)
        merge_sort(right)

        i=j=k=0
        while i<len(left) and j<len(right):
            if left[i]>right[j]:
                arr[k]=right[j]
                k+=1
                j+=1
            else:
                arr[k]=left[i]
                i+=1
                k+=1
        while i<len(left):
            arr[k]=left[i]
            i+=1
            k+=1
        while j<len(right):
            arr[k]=right[j]
            j+=1
            k+=1
arr=[2,5,3,1,4,7]
merge_sort(arr)
print(arr)
```
