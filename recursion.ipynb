```python
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
print(factorial(5))
```

```python
def fibonacci(n):
    if n<=1:
        return n
    return fibonacci(n-1)+fibonacci(n-2)
print(fibonacci(6))
```

```python
def sum(n):
    if n==0:
        return 0
    return n%10+sum(n//10)
print(sum(123))
```

```python
def reverse(s):
    if len(s)==0:
        return ""
    return s[-1]+reverse(s[:-1])
print(reverse("hello"))
```

```python
def is_palindrome(s):
    if len(s) <= 1:
        return True
    return s[0] == s[-1] and is_palindrome(s[1:-1])
print(is_palindrome("malayalam"))
```

```python
def sumofnumbers(n):
    if n==0:
        return 0
    
    return n+sumofnumbers(n-1)
print(sumofnumbers(10))
```

```python
def binary_search(arr,low,high,target):
        if low>high:
             return-1
        mid=(low+high)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            return  binary_search(arr,mid+1,high,target)
        else:
            return binary_search(arr,low,mid-1,target)
        
arr=[1,2,3,4,5,6,7]
print(binary_search(arr,0,len(arr)-1,5))
```

```python
def pow(a,b):
    if b==0:
        return 1
    return a*pow(a,b-1)
print(pow(2,3))
```

```python
def tower_of_hanoi(n,source,helper,destination):
    if n==1:
        print(f"move {n} from {source} to {destination}")
        return
    tower_of_hanoi(n-1,source,destination,helper)
    print(f"move {n} from {source} to {destination}")
    tower_of_hanoi(n-1,helper,source,destination)

tower_of_hanoi(3,'A','B','C')
```

```python
def printnreverse(n):
    if n==0:
        return 0
    print(n)
    printnreverse(n-1)
printnreverse(6)
```

```python
def len_of_list(arr):
    if arr==[]:
        return 0
    return 1+len_of_list(arr[1:])
print(len_of_list([1,2,3,4]))
```

```python
def to_binary(n):
    if n==0:
        return ''
    return to_binary(n//2) +str(n%2)
print(to_binary(10))
```

```python
def replace_char(s,old,new):
    if not s:
        return ""
    if s[0]==old:
        return new+replace_char(s[1:],old,new)
    else:
        return s[0]+replace_char(s[1:],old,new)
print(replace_char("hello",'l','w'))
```
