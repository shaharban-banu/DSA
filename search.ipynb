```python
#binary search
arr = [1, 3, 5, 7, 9, 11,3]
target = 9
low,high=0,len(arr)-1
while low<=high:
    mid=(low+high)//2
    if arr[mid]==target:
        print(f"found at {mid}")
        break
    elif arr[mid]<target:
        low=mid+1
    else :
        high=mid-1

```

```python
#. First and Last Occurrence by binary search
arr = [2, 4, 4, 4, 6, 8]
target = 4

def first_occurence(arr,target):
    low,high,result=0,len(arr)-1,-1
    while low<=high:
        mid=(low+high)//2
        if arr[mid]==target:
            result=mid
            high=mid-1
        elif arr[mid]<target:
            low=mid+1
        else:
            high=mid-1
    return result

def last_occurence(arr,target):
    low,high,r=0,len(arr)-1,-1
    while low<=high:
        mid=(low+high)//2
        if arr[mid]==target:
            r=mid
            low=mid+1
        elif arr[mid]<target:
            low=mid+1
        else:
            high=mid-1
    return r
print(first_occurence(arr,target))
print(last_occurence(arr,target))
```
