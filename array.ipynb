```python
arr=[1,4,3,2,5]
k=2
n=len(arr)
k=k%n
def reverse(arr,left,right):
    while left<right:
        arr[left],arr[right]=arr[right], arr[left]
        left+=1
        right-=1
    return arr
reverse(arr,0,n-1)
reverse(arr,0,k-1)
reverse(arr,k,n-1)
print(arr)
```

```python
nums = [5, 7, 7, 8, 8, 10]
target = 8

def first_occurence(nums,target):
    n=-1
    low=0
    high=len(nums)-1
    while low<=high:
        mid =(low+high)//2
        if nums[mid]==target:
            n=mid
            high=mid-1
        elif nums[mid]<target:
            low=mid+1
        else:
            high=mid-1

    return n

def last_occurence(nums,target):
    n=-1
    low=0
    high=len(nums)-1
    while low<=high:
        mid =(low+high)//2
        if nums[mid]==target:
            n=mid
            low=mid+1
        elif nums[mid]<target:
            low=mid+1
        else:
            high=mid-1
            
    return n

print(first_occurence(nums,target))
print(last_occurence(nums,target))
```

```python
res=[]
def backtrack(path,used):
    if len(path)==len(nums):
        res.append(path[:])
        return
    for i in range(len(nums)):
        if not used[i]:
            used[i]=True
            path.append(nums[i])
            backtrack(path,used)
            path.pop()
            used[i]=False
backtrack([],[False]*len(nums))
print(res)
```
