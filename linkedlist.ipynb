```python
class Node:
    def __init__(self, data):
        self.data = data
        self.ref = None  # using 'ref' instead of 'next'

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            n = self.head
            while n.ref is not None:
                n = n.ref
            n.ref = new_node

    def print_list(self):
        n = self.head
        while n is not None:
            print(n.data, end=" -> ")
            n = n.ref
        print("None")

    def remove_duplicate(self):
        n = self.head
        while n is not None and n.ref is not None:
            if n.data == n.ref.data:
                # Skip the duplicate node
                n.ref = n.ref.ref
            else:
                n = n.ref

# ---------- Test ----------
ll = LinkedList()
for val in [10, 20, 20, 30, 30,30, 40, 40, 40, 50,50]:
    ll.append(val)

print("Original List:")
ll.print_list()

ll.remove_duplicate()

print("After Removing Duplicates:")
ll.print_list()
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.nref=None
        self.pref=None

class Doublyll:
    def __init__(self):
        self.head=None
    
    def insert_empty(self,data):
        if self.head is None:
            new_node=Node(data)
            self.head=new_node
        else:
            print("ll is not empty")

    def print(self):
        if self.head is None:
            print("empty")
        else:
            n=self.head
            while n is not None:
                print(n.data,end="->")
                n=n.nref
            print("none")

    def insert_begin(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            new_node.nref=self.head
            self.head.pref=new_node
            self.head=new_node
    
    def insert_end(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
        else:
            n=self.head
            while n.nref is not None:
                n=n.nref
            n.nref=new_node
            new_node.pref=n

    def insert_after(self,data,x):
        new_node=Node(data)
        if self.head is None:
            print("empty")
        else: 
            n=self.head
            while n is not None:
                if n.data==x:
                    break
                n=n.nref
            if n is None:
                print("key not found")
            else:
                new_node.nref=n.nref
                if n.nref is not None:
                    n.nref.pref=new_node
                new_node.pref=n
                n.nref=new_node

    def insert_before(self,data,x):
        new_node=Node(data)
        if self.head is None:
            print("empty")
        else:
            n=self.head
            while n.nref is not None:
                if n.nref.data==x:
                    break
                n=n.nref
            if n.nref is None:
                print("key not found")
            else:
                new_node.nref=n.nref
                n.nref.pref=new_node
                n.nref=new_node
                new_node.pref=n

    def delete_begin(self):
        if self.head is None:
            print("deletion not possible")
        elif self.head.nref is None:
            self.head=None
        else:
            self.head=self.head.nref
            self.head.pref=None

    def delete_end(self):
        if self.head is None:
            print("deletion not possible")
        elif self.head.nref is None:
            self.head=None
        else:
            n=self.head
            while n.nref is not None:
                n=n.nref
            n.pref.nref=None

    def delete_by_value(self,x):
        if self.head is None:
            print("deletion not possible")
        elif self.head.nref is None:
            if self.head.data==x:
                self.head=None
            else:
                print("key not found")
        elif self.head.data==x:
            self.head=self.head.nref
            self.head.pref=None
        else:
            n=self.head
            while n.nref is not None:
                if x==n.data:
                    break
                n=n.nref
            if n.nref is not None:
                n.pref.nref=n.nref
                n.nref.pref=n.pref
            else:
                n.pref.nref=None
                  


d=Doublyll()
d.insert_empty(10)
d.insert_begin(20)
d.insert_end(50)
d.insert_after(60,20)
d.insert_after(70,50)
d.insert_before(11,10)
d.print()

d.delete_begin()
d.delete_end()
d.delete_by_value(10)
d.print()

```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1, l2):
        dummy = ListNode()
        current = dummy
        carry = 0

        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0

            total = val1 + val2 + carry
            carry = total // 10

            current.next = ListNode(total % 10)
            current = current.next

            if l1: l1 = l1.next
            if l2: l2 = l2.next

        return dummy.next

# Helpers
def create_linked_list(arr):
    dummy = ListNode()
    current = dummy
    for num in arr:
        current.next = ListNode(num)
        current = current.next
    return dummy.next

def print_linked_list(head):
    while head:
        print(head.val, end=" -> ")
        head = head.next
    print("None")

# Create input lists
l1 = create_linked_list([2, 4, 3])  # 342
l2 = create_linked_list([5, 6, 4])  # 465

# Solve
sol = Solution()
result = sol.addTwoNumbers(l1, l2)

# Output
print_linked_list(result)  # Output: 7 -> 0 -> 8 -> None
```

```python
#sorting
class node:
    def __init__(self, data):
        self.data = data
        self.ref = None

class solution:
    def __init__(self):
        self.head = None

    def insert_end(self, data):
        new_node = node(data)
        if not self.head:
            self.head = new_node
        else:
            n = self.head
            while n.ref:
                n = n.ref
            n.ref = new_node

    def printll(self):
        if not self.head:
            print("empty")
        else:
            n = self.head
            while n:
                print(n.data, end="-->")
                n = n.ref
            print("none")

    def merge_sort(self):
        def _merge_sort(head):
            if not head or not head.ref:
                return head
            mid=self._mid(head)
            next_mid=mid.ref
            mid.ref=None

            left=_merge_sort(head)
            right=_merge_sort(next_mid)

            return self.sorted_merge(left,right)
        self.head=_merge_sort(self.head)


    def _mid(self,head):
        slow=head
        fast=head.ref
        while fast and fast.ref:
            slow=slow.ref
            fast=fast.ref.ref
        return slow 
    
    def sorted_merge(self,a,b):
        if not a:
            return b
        if not b:
            return a
        if a.data<b.data:
            result=a
            result.ref=self.sorted_merge(a.ref,b)
        else:
            result=b
            result.ref=self.sorted_merge(a,b.ref)
        return result


arr = [5, 2, 9, 1, 6, 3]
s = solution()
for item in arr:
    s.insert_end(item)

s.printll()

s.merge_sort()
s.printll()

```
